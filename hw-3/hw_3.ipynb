## $ \textbf{III.5.12в,г*} $


### Условие:
  Согласно переписи население США менялось следующим образом:
  * 1910 – 92 228 496 человек,
  * 1920 – 106 021 537,
  * 1930 – 123 202 624,
  * 1940 – 132 164 569,
  * 1950 – 151 325 798,
  * 1960 – 179 323 175,
  * 1970 – 203 211 926,
  * 1980 – 226 545 805,
  * 1990 – 248 709 873,
  * 2000 – 281 421 906.

  Используя многочлены степеней N = 2, 3, 4, 5 построить аппроксимацию
этих данных в смысле МНК. Аппроксимацию можно строить на основе базисных многочленов следующих видов. \\
  
  Какое представление базисных многочленов является наилучшим? \\
  
  Используйте построенные приближения для предсказания численно-
сти населения США в 2010 году и сравните с точным результатом
в 308 745 538 человек.



### Решение:
  * $ \textbf{в)} $ \\
    $$ f(x) ≈ ∑_{0}^{N} c_n(x-1955)^n $$

    N=2: \\
    $$ f(x) ≈ c_0 + c_1(x-1955) + c_2(x-1955)^2 $$

    N=3: \\
    $$ f(x) ≈ c_0 + c_1(x-1955) + c_2(x-1955)^2 + c_3(x-1955)^3 $$

    N=4: \\
    $$ f(x) ≈ c_0 + c_1(x-1955) + c_2(x-1955)^2 + c_3(x-1955)^3 + c_4(x-1955)^4$$

    N=5: \\
    $$ f(x) ≈ c_0 + c_1(x-1955) + c_2(x-1955)^2 + c_3(x-1955)^3 + c_4(x-1955)^4 + c_5(x-1955)^5 $$

    Cреднеквадратическую ошибка: \\
    $$ ρ(y, g) = (∑_{1}^{10}\frac{1}{10} \cdot |y(i)-g(i)|^2)^{1/2} $$ \\

    Решение в wolfram в pdf файле. \\

    В pdf по отдельности расмотрены разные N. \\

    Выводимые значения: \\
    - Коэффициенты
    - Cреднеквадратическая ошибка
    - Разность модуля предсказания и настоящего значения на 2010
    - График

    $ \textbf{Тут будут выводы}: $
    - Мин. ошибка при: $ N = 5 → 1.78692 \cdot 10^6 $.
    - Лучшее предсказание при: $ N = 3 \rightarrow 275441 - модуль \ разности $.
